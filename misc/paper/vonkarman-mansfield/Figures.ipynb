```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
r0 = 1.
t0 = 1e-2
nu = 0.3
R0 = r0**2/t0
c = 2/(1+nu)*np.sqrt(2*(7+nu)/(1+nu))
prsinagdhi5 = np.loadtxt("output/data-psri-naghdi-5.csv")
plt.plot(prsinagdhi5[0]*R0,prsinagdhi5[1]*R0,"-.r",label="Naghdi")
plt.plot(prsinagdhi5[0]*R0,prsinagdhi5[2]*R0,"-.r")
prsi5 = np.loadtxt("output/data-psri-5-large.csv")
plt.plot(prsi5[0]*R0,prsi5[1]*R0,"-g",label = "Von Karman")
plt.plot(prsi5[0]*R0,prsi5[2]*R0,"-g")
plt.xlim([0,2])
plt.ylim([-1,2])
plt.xlabel(r"inelastic curvature $\bar\kappa \,a^2/t$")
plt.ylabel(r"average curvature $k\,a^2/t$")
plt.legend()
plt.tight_layout()
plt.savefig("naghdi-prsi-comparison.pdf")
plt.xlim([0,50])
plt.ylim([-5,50])
plt.tight_layout()
plt.savefig("naghdi-prsi-comparison-large.pdf")
```

```python
r0 = 1.
t0 = 1e-2
R0 = r0**2/t0
#c_cr = 5.16/R0
# Analytical solution
h_before = np.loadtxt("mansfield-solution/mansfield_ktpre.csv")
h_after = np.loadtxt("mansfield-solution/mansfield_ktpost.csv")
ls_Kbefore = np.loadtxt("mansfield-solution/mansfield_kpre.csv")
ls_K1after = np.loadtxt("mansfield-solution/mansfield_kxpost.csv")
ls_K2after = np.loadtxt("mansfield-solution/mansfield_kypost.csv")

# Comparison
fig = plt.figure(figsize=(5.0, 5.0/1.648))
plt.plot(h_before, ls_Kbefore, "-", color='b', label="Analytical solution $k_{1}$, $k_{2}$")
plt.plot(h_after, ls_K1after, "-", color='b')
plt.plot(h_after, ls_K2after, "-", color='b')
plt.xlabel(r"inelastic curvature $\bar\kappa \,L^2/t$")
plt.ylabel(r"average curvature $k\,L^2/t$")
plt.tight_layout()
#plt.savefig("MITC-curvature_bifurcation.png")
plt.plot(prsi5[0]*R0,prsi5[1]*R0,"xg", label = "PRSI: n = 5")
plt.plot(prsi5[0]*R0,prsi5[2]*R0,"xg")
plt.plot(prsi10[0]*R0,prsi10[1]*R0,".r", label = "PRSI: n = 10")
plt.plot(prsi10[0]*R0,prsi10[2]*R0,".r")
plt.xlim([0,8])
plt.ylim([0,9])
plt.legend()
plt.savefig("comparison-prsi.eps")
plt.savefig("comparison-prsi.pdf")
plt.show()
```
